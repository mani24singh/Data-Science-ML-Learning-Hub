## Handling Imbalanced Dataset

1. Up Sampling: If we have a lot of missing values in the data, we can use up sampling to fill the missing values. We can use the mean, median, or mode of the data to fill the missing values. 
2. Down Sampling: If we have a lot of missing values in the data, we can use down sampling to fill the missing values. We can use the mean, median, or mode of the data to fill the missing values.

```python
 import pandas as pd
 import numpy as np
```

```python
# Set the random seed for reproducibility
np.random.seed(42)

# Create a dataframe with two classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0, n_class_1
```

```python
# Create my dataframe with imbalanced dataset
class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0), 
    'target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1), 
    'target': [1] * n_class_1
})
```

```python
df=pd.concat([class_0,class_1]).reset_index(drop=True)
```

```python
df.tail()
```

```python
df.target.unique()
```

```python
df['target'].value_counts()
```

### Up-Sampling

```python
df_minority = df[df['target']==1]
df_majority = df[df['target']==0]
```

```python
from sklearn.utils import resample

df_minority_upsampled = resample(df_minority, replace=True, # sample with replacement
                                 n_samples=len(df_majority), # match number in majority class
                                 random_state=42) # reproducible results    
```

```python
df_minority_upsampled.shape
```

```python
df_minority_upsampled.head()
```

```python
df_upsampled = pd.concat([df_majority, df_minority_upsampled])
```

```python
df_upsampled['target'].value_counts()
```

### Down-Sampling

```python
import pandas as pd

# Set the random seed for reproducibility
np.random.seed(123)

# Create a dataframe with two classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0

class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1] * n_class_1
})

df = pd.concat([class_0, class_1]).reset_index(drop=True)

# Check the class distribution
print(df['target'].value_counts())
```

```python
# Downsampling
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
from sklearn.utils import resample

df_majority_downsampled=resample(df_majority,replace=False,
         n_samples=len(df_minority), # length of minority class
         random_state=42
        )
```

```python
df_majority_downsampled.shape
```

```python
df_downsampled = pd.concat([df_minority, df_majority_downsampled])
```

```python
df_downsampled.target.value_counts()
```

- Note: 
1. Down Sampling is not recommended when you have a small dataset as it can lead to loss of information. It is generally used when you have a large dataset and want to reduce the training time of your model.
2. Down sampling is bad compared to up sampling because it can lead to loss of information. Up sampling is generally preferred when you have a small dataset and want to increase the number of samples in the minority class.
